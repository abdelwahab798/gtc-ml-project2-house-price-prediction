# Data Cleaning

```python
df["longitude"]=df["longitude"].abs()
```

```python
df["total_bedrooms"] = df["total_bedrooms"].fillna(df.groupby("ocean_proximity")["total_bedrooms"].transform("median"))
```

# Feature Engineering

```python
df["width&lenght"]=df["longitude"]*df["latitude"]
df["rooms_without_bedrooms"]=df["total_rooms"]-df["total_bedrooms"]
df["population_per_household"]=df["population"]/df["households"]
df["bedrooms_per_room"]=df["total_bedrooms"]/df["total_rooms"]
```

```python
df = pd.get_dummies(df, columns=["ocean_proximity"], drop_first=True)
```

# Modelling

```python
from sklearn.model_selection import train_test_split
x=df.drop("median_house_value", axis=1)
y=df["median_house_value"]
x_trian, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=42)
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
model_lin=LinearRegression()
model_lin.fit(x_trian, y_train)
y_pred=model_lin.predict(x_test)
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

# Xgb regressor

```python
from xgboost import XGBRegressor
model_xgb=XGBRegressor(n_estimators=500,learning_rate=0.05,max_depth=6,subsample=0.8,colsample_bytree=0.8,random_state=42)
model_xgb.fit(x_trian, y_train)
y_pred=model_xgb.predict(x_test)
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
importances= model_xgb.feature_importances_
features= x.columns
plt.figure(figsize=(10,6))
plt.barh(features, importances)
plt.xlabel("Feature Importance")
plt.title("XGBoost Feature Importances")
feature_importance = pd.Series(importances, index=features).sort_values(ascending=False)
print(feature_importance)
```
